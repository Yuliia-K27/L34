1. Створити xls-документ, що містить дані про міста, їх населення, середній дохід жителів, площу території та наявність метрополітену.
Кількість записів у документі - 20.

```python
import pandas as pd
import matplotlib.pyplot as plt

cities = {
    'city_name': ['Kyiv', 'Lviv', 'Odesa', 'Kharkiv', 'Dnipro', 'Zaporizhzhya', 'Chernihiv', 'Cherkasy', 'Sumy', 'Zhytomyr', 'Ternopil', 'Ivano-Frankivsk', 'Lutsk', 'Kropyvnytsky', 'Mykolaiv', 'Kherson', 'Chernivtsy', 'Ryvne', 'Uzhhorod','Kryvyi Rih'],
    'city_population': [2956582, 721301, 993120, 1445000, 1008567, 747929, 293827, 280413, 264325, 270918, 228505, 230038, 206802, 240883, 494363, 307650, 278023, 246338, 117317, 660115],
    'average_income': [32000, 23000, 21000, 20000, 19000, 17000, 15000, 14000, 13000, 12000, 14000, 18000, 19000, 18000, 17000, 16000, 15000, 14000, 13000, 20000],
    'city_area': [839, 182, 162, 310, 405, 334, 153, 69, 120, 65, 43, 84, 41, 52, 260, 331, 152, 24, 77, 307],
    'subway': ['yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'no', 'yes'],
}

# Create the dataframe.
df = pd.DataFrame(data=cities, index=list(range(0, 20)))
print('cities:')
print(df)
```

```python
# Save the dataframe to xlsx-file.
df.to_excel('cities.xlsx', index=False)
```

2. За допомогою plotly візуалізувати залежність населення міста від його площі та середнього доходу населення. Помістити ці графіки на один малюнок.

```python
# Create the plot with two subplots.
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plot the dependency of city population on its area.
ax[0].scatter(df['city_area'], df['city_population'], alpha=0.5)
ax[0].set_xlabel('City area (km^2)')
ax[0].set_ylabel('City population')
ax[0].set_title('Dependency of city population on its area')

# Plot the dependency of city population on its average income.
ax[1].scatter(df['average_income'], df['city_population'], alpha=0.5)
ax[1].set_xlabel('Average income')
ax[1].set_ylabel('City population')
ax[1].set_title('Dependency of city population on its average income')

# Show the plot.
plt.show()
```

3. Побудувати матрицю кореляцій та візуалізувати її через Heatmap.

```python
import seaborn as sns

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```
